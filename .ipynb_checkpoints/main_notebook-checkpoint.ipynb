# **Unsupervised ECG Interpretation**

-------

## Goal

*  Design and implement an ML-based automatic electrocardiogram interpretation algorithm in python.

*   The aim is to partition electrocardiogram signals into distinct classes without cardiologist-supplied labels.

*   The supplied signal labels are only to be used during testing and evaluation, not during model training.

*   use the [PTB-XL dataset available on PhysioNet](https://www.physionet.org/content/ptb-xl/1.0.3/)



## Chosen Approach

I am not too familiar with cardiology. That is why I choose to follow the following data-driven, perhaps a bit naif approach:

1. I will train an **autoencoder** to reconstruct the input ECG signals.
2. I will then extract the **encoded representations** obtained from the trained autoencoder; this way I will bypass the need of having to compute and select features by hand, as the encoded latent representations of a (hopefully! ðŸ€) performant autoencoder will by definition capture meaningful info about the input signal.
3. I will use the obtained encoded representations as input features to perform  **K-Means** and **Hierarchical Clustering**, plotting a dendrogram and search for a threshold to use to group them into clusters.

### Diagnostic Superclass vs. Age, Sex and BMI

To make the EDA even easier and clearer, I will focus only on these ECG records having only one diagnostic_superclass. I will also keep only records having both `weight` and `height`, from which I will calculate the body-mass index (BMI), which may be associated with cardiovascular health.

```python
def keep_only_unique_diagnostic_superclasses(df):

  # Apply the function to the diagnostic superclass column
  df['diagnostic_superclass'] = df['diagnostic_superclass'].apply(filter_diagnostic)

  # Filter out rows where diagnostic superclass is None
  df.dropna(subset=['diagnostic_superclass'], inplace=True)

  return df


def filter_diagnostic(diagnostic_str):
  diagnostic_to_keep = ['NORM', 'MI', 'STTC', 'CD', 'HYP']

  if diagnostic_str in diagnostic_to_keep:
    return diagnostic_str

  else:
    try:
        # Evaluate the string to a list
        diagnostic_list = ast.literal_eval(diagnostic_str)
        # Check if it's a single-element list and if the element is in the valid list
        if len(diagnostic_list) == 1 and diagnostic_list[0] in diagnostic_to_keep:
            return diagnostic_list[0]
        else:
            return None
    except:
        return None
```

```python
# simplify EDA: keep only records having one diagnostic_superclass
meta_redux = keep_only_unique_diagnostic_superclasses(meta)

# keep only subjects having 110 years max (remove 300 years old outliers ...)
meta = meta_redux.query("age < 110")

# select only relevant variables to keep the dataframe lean
vars_to_keep = ['patient_id', 'diagnostic_superclass', 'diagnostic_label',
                'age', 'sex', 'weight', 'height', 'filename_lr']

# keep only those records having both `weight` and `height`, compute BMI
meta_redux = meta[vars_to_keep].dropna(subset=['weight', 'height'])
meta_redux['BMI'] = meta_redux['weight'] / (meta_redux['height'] / 100) ** 2
```

# 2. Model Training

### Helper Functions


```python
def parse_signal_and_signal(example):
  """
  Parse a TFRecord example into signal and signal tensors,
  needed as input for train an autoencoder.

  Args:
      proto: A serialized tf.train.Example.

  Returns:
      signal: A tf.Tensor representing the signal data (float32).
      signal: A tf.Tensor representing the signal data (float32).
  """

  feature_description = {
      'signal': tf.io.VarLenFeature(tf.float32),
  }
  parsed_features = tf.io.parse_single_example(example, feature_description)
  # Decode and reshape the signal and label appropriately.
  signal = tf.sparse.to_dense(parsed_features['signal'])
  signal = tf.reshape(signal, shape=[1000, 12])

  return signal, signal



def parse_signal_and_label(example):
  """
  Parse a TFRecord example into signal and label tensors.
  Useful for test the trained autoencoder.

  Args:
      proto: A serialized tf.train.Example.

  Returns:
      signal: A tf.Tensor representing the signal data (float32).
      label: A tf.Tensor representing the label (string).
  """

  feature_description = {
      'signal': tf.io.VarLenFeature(tf.float32),
      'label': tf.io.FixedLenFeature([], dtype=tf.string),
  }
  parsed_features = tf.io.parse_single_example(example, feature_description)
  # Decode and reshape the signal and label appropriately.
  signal = tf.sparse.to_dense(parsed_features['signal'])
  signal = tf.reshape(signal, shape=[1000, 12])

  label = parsed_features['label']

  return signal, label


def compute_mean_std(dataset, samples=10000):
    # Initialize sum and square sum to compute mean and std
    sum1 = 0
    sumsq = 0
    count = 0

    for signal, _ in dataset.take(samples):
        sum1 += tf.reduce_sum(signal, axis=0)
        sumsq += tf.reduce_sum(tf.square(signal), axis=0)
        count += signal.shape[0]

    mean = sum1 / count
    variance = (sumsq / count) - tf.square(mean)
    std = tf.sqrt(variance)

    return mean, std


def create_normalized_training_dataset(tfrecords_dir, batch_size, num_samples=None):
    n_samples = 21799 if num_samples is None else num_samples
    tfrecord_files = tf.data.Dataset.list_files(os.path.join(tfrecords_dir, "*.tfrec"))
    raw_dataset = tf.data.TFRecordDataset(tfrecord_files)
    parsed_dataset = raw_dataset.map(parse_signal_and_signal)

    # Compute mean and std from a subset of the data
    mean, std = compute_mean_std(parsed_dataset, samples=n_samples)

    # Function to normalize data
    def normalize_data(signal, signal_output):
        return (signal - mean) / std, (signal_output - mean) / std

    # Apply normalization and prepare dataset
    normalized_dataset = parsed_dataset.map(normalize_data)
    normalized_dataset = normalized_dataset.shuffle(buffer_size=n_samples).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

    return normalized_dataset


def create_normalized_test_dataset(tfrecords_dir, batch_size, num_samples=None):
    n_samples = 21799 if num_samples is None else num_samples
    tfrecord_files = tf.data.Dataset.list_files(os.path.join(tfrecords_dir, "*.tfrec"))
    raw_dataset = tf.data.TFRecordDataset(tfrecord_files)
    parsed_dataset = raw_dataset.map(parse_signal_and_label)

    # Compute mean and std from a subset of the data
    mean, std = compute_mean_std(parsed_dataset, samples=n_samples)

    # Function to normalize data
    def normalize_data(signal, label):
        return (signal - mean) / std, label

    # Apply normalization and prepare dataset
    normalized_dataset = parsed_dataset.map(normalize_data)
    normalized_dataset = normalized_dataset.shuffle(buffer_size=n_samples).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

    return normalized_dataset
```

### Model building Function

```python
def build_autoencoder(input_shape=(1000, 12)):

    model_name = 'CONV1D'

    # Define the input tensor
    input_layer = layers.Input(shape=input_shape)  # shape: (None, 1000, 12)

    # Encoder
    x = layers.Conv1D(64, 3, activation='relu', padding='same')(input_layer)  # shape: (None, 1000, 64)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)  # shape: (None, 500, 64)

    x = layers.Conv1D(32, 3, activation='relu', padding='same')(x)  # shape: (None, 500, 32)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)  # shape: (None, 250, 32)

    x = layers.Conv1D(16, 3, activation='relu', padding='same')(x)  # shape: (None, 250, 16)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(5)(x)  # shape: (None, 50, 16)

    x = layers.Conv1D(8, 3, activation='relu', padding='same', 
                      activity_regularizer=regularizers.l1(0.01))(x)  # shape: (None, 50, 8)
    x = layers.BatchNormalization()(x)
    encoded = layers.MaxPooling1D(2, name='bottleneck_layer')(x)  # shape: (None, 25, 8)

    # Decoder
    x = layers.Conv1DTranspose(8, 3, activation='relu', strides=2, padding='same')(encoded) # shape: (None, 50, 8)
    x = layers.BatchNormalization()(x)

    x = layers.Conv1DTranspose(16, 3, activation='relu', strides=5, padding='same')(x) # shape: (None, 250, 16)
    x = layers.BatchNormalization()(x)

    x = layers.Conv1DTranspose(32, 3, activation='relu', strides=2, padding='same')(x) # shape: (None, 500, 32)
    x = layers.BatchNormalization()(x)

    x = layers.Conv1DTranspose(64, 3, activation='relu', strides=2, padding='same')(x) # shape: (None, 1000, 64)
    x = layers.BatchNormalization()(x)

    x = layers.Conv1DTranspose(12, 3, activation='relu', padding='same')(x)  # shape: (None, 1000, 12)
    x = layers.BatchNormalization()(x)

    autoencoder = Model(inputs=input_layer, outputs=x, name=model_name)

    return autoencoder, model_name
```

### Build and Compile Autoencoder

```python
autoencoder, model_name = build_autoencoder()
autoencoder.summary()
```

```python
autoencoder.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=0.001, weight_decay=0.0001),
    loss='mse'
)
```

```python
# Define callbacks
reduce_lr_on_plateau = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',    # Monitor loss
    factor=0.5,            # Reduce learning rate by a factor of 0.5
    patience=5,            # Number of epochs with no improvement before reducing the learning rate
    min_lr=1e-6            # Minimum learning rate
)

early_stopping = EarlyStopping(
    monitor='val_loss',  # Metric to monitor
    patience=15,         # Number of epochs with no improvement after which training will be stopped
    verbose=1,           # If 1, prints a message when stopping
    restore_best_weights=True  # Restores model weights from the epoch with the best value of the monitored metric
)
```

### Plot Original and Reconstructed ECG signal

```python
def plot_ecg_comparison(original_data, reconstructed_data, model_name=None, num_samples=1000):
    """
    Function to plot comparison of original and reconstructed ECG data.

    Parameters:
    original_data (numpy array): The original ECG data. Expected shape: (num_samples, num_channels)
    reconstructed_data (numpy array): The reconstructed ECG data. Expected shape: (num_samples, num_channels)
    num_samples (int): Number of samples in each channel. Default is 1000.
    """
    num_channels = original_data.shape[1]
    time = np.linspace(0, num_samples / 100, num_samples)  # 100 Hz sampling rate, so 100 samples per second

    fig, axs = plt.subplots(num_channels, 1, figsize=(7, 10), sharex=True)
    for i in range(num_channels):
        axs[i].plot(time, original_data[:, i], linewidth=1.5, color='lightgray')
        axs[i].plot(time, reconstructed_data[:, i], linewidth=1, color='tab:orange')
        axs[i].set_xticklabels([])  # Remove x-tick labels

    # Set the x-axis label for the last subplot
    axs[-1].set_xlabel("time (s)", fontsize=10)

    # Adjust layout to be tighter
    model_name = 'unnamed' if model_name is None else model_name
    suptitle = f"Original (gray) vs. Reconstructed (orange) ECG (AE model = '{model_name}')"
    plt.suptitle(suptitle, fontsize=10, fontweight='bold')
    plt.tight_layout()
    plt.subplots_adjust(hspace=0.1, top=0.95, bottom=0.05, left=0.05, right=0.95)
    plt.show()


def extract_random_instance(dataset):
    """
    Extract a random ECG data instance from the dataset.

    Parameters:
    dataset (tf.data.Dataset): The dataset from which to extract the ECG data instance.

    Returns:
    numpy array: A single ECG data instance.
    """
    # Take one batch from the dataset
    for batch in dataset.take(1):
        ecg_data, _ = batch  # Extract the first element of the batch
        break

    # Select a random instance from the batch
    random_index = tf.random.uniform(shape=[], minval=0, maxval=tf.shape(ecg_data)[0], dtype=tf.int32)
    ecg_instance = ecg_data[random_index].numpy()

    return ecg_instance
```

### Cluster the Encoded ECG Features

In this section we evaluate:

1.   if the encoded ECG features could be used to group ECG signals into separate groups and
2.   based on the clustering results, we will try to estimate the optimal number of clusters into which to split all ECG signals.

To do so, we will apply and analyze the results of 2 types of clustering techniques,  **K-Means** and **Hierarchical Clustering**.

#### K-Means

The "Elbow Method"  involves plotting the **inertia** (i.e. the *within-cluster sum of squares*) as a function of the number of clusters, and picking the elbow of the curve as the number of clusters to use. This method is intuitive and widely used but can sometimes be ambiguous if a clear "elbow" is not present.

The **silhouette score** measures how similar an object is to its own cluster compared to other clusters. The silhouette scores range from -1 to +1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.

Inertia (using the Elbow method) and Silhouette score can be used, together, to estimate the optimal number of clusters into which to split the data.

```python
inertia = []
silhouette_scores = []
for k in range(2, 11):  # Silhouette score cannot be computed with a single cluster
    print(f"KMEANS: compute inertia and silhouette score for k = {k} clusters")
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(reshaped_encoded_data)
    inertia.append(kmeans.inertia_)
    score = silhouette_score(reshaped_encoded_data, kmeans.labels_)
    silhouette_scores.append(score)
```

```python
# Plot Inertia and Silhouette Scores per number of clusters
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(range(2, 11), inertia, marker='o')
plt.ylabel('Inertia', fontweight='bold')

plt.subplot(2, 1, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score', fontweight='bold')

# Get current x-axis ticks locations for all subplots
for ax in plt.gcf().get_axes():
  ax.xaxis.set_major_locator(plt.FixedLocator(range(2, 20)))

# Filter only integer tick labels for all subplots
for ax in plt.gcf().get_axes():
  ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: f"{int(val)}" if val.is_integer() else ''))

plt.show()
```

Both the silhouette score and the Inertia have "bad" shapes, meaning neither of them clearly suggest a potential optimal number of cluster.  Ouch ðŸ¤”

Therefore I will try to use Hierarchical Clustering and see If we can get a more conclusive split.

#### Hierarchical Clustering

The optimal number of clusters can be estimated with the Elbow Criterion also from the hierarchical cluster's **scree plot**, which represent the cluster distance dividing each next "cluster merge".  



```python
# get linkage matrix and clusters merge distances vector (this take time!)
hc_linkage_matrix = linkage(reshaped_encoded_data, metric="euclidean", method='ward')
clusters_merge_distances = hc_linkage_matrix[:, 2]

# find optimal number of clusters using the "elbow" criterion:
acceleration = np.diff(clusters_merge_distances, 2)  # 2nd derivative of the distances
acceleration_flipped = np.flip(acceleration)
hc_optimal_n_of_clusters = acceleration_flipped.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
branch_height = clusters_merge_distances[-hc_optimal_n_of_clusters]  # lowest distance before last merge
cutting_distance = branch_height + (branch_height / 100 * 1)  # to ensure to cut at the right distance

print("HC Estimated Optimal Number of Clusters: ", hc_optimal_n_of_clusters)
```

```python
plt.figure(figsize=(20, 7))

# PLOT DENDROGRAM
ax1 = plt.subplot(1, 2, 1)
dendrogram(hc_linkage_matrix, orientation='top', distance_sort='descending', show_leaf_counts=True, truncate_mode='lastp')
plt.axhline(cutting_distance, ls=':')

# Update the tick labels to keep only those within parentheses
xticks = ax1.get_xticklabels()
new_labels = []
for label in xticks:
    text = label.get_text()
    if '(' in text:
        new_labels.append(text)
    else:
        new_labels.append('')

# Set the new tick labels
ax1.set_xticklabels(new_labels)
plt.xlabel('number of observations per branch'), plt.ylabel('distance between clusters')
plt.title('Hierarchical Clustering Dendrogram', fontweight='bold')


# SCREE PLOT:
ax2 = plt.subplot(1, 2, 2)
xticklabels = [x for x in range(1, len(clusters_merge_distances))]
xticks = [x for x in range(len(clusters_merge_distances) - xticklabels[-1], len(clusters_merge_distances))]
plt.plot(clusters_merge_distances, '-o', ms=3, mfc='white')
plt.axhline(cutting_distance, ls=':')
plt.xticks(ticks=xticks[::10], labels=xticklabels[::-1][::10])
plt.xlim([xticks[-100], xticks[-1]])
plt.xlabel('num. of remaining clusters')
plt.ylabel("distance between merged clusters")
plt.title('Scree Plot', fontweight='bold')

plt.show()
```

The HC results suggests the existence of 4 potential clusters, not sodifferent from the true 5 ...

### Evaluate Clustering Based on Encoded Representations

Here we quantify the relationship between the predicted unsupervised clusters assigned to the ECG signals and the true diagnostic class, assuming the suggested "optimal" number of groups into which the data should be split according to HC and K-Means.

We will evaluate such relationship using:

1.   **Confusion Matrix**: This helps you see how the clusters correspond to the true labels. A confusion matrix will show the distribution of true labels across the clusters.

2.   **Adjusted Rand Index (ARI)**: This measures the similarity between the true labels and the clustering results, adjusting for chance. The ARI ranges from -1 to 1, being 1 perfect agreement between the true labels and the clustering, 0 random labeling, and negative values worst the random clustering.

3.   **Normalized Mutual Information (NMI)**: This measures the amount of shared information between the true labels and the clusters. NMI ranges from 0 to 1, where 1 stands for perfect correlation between the true labels and the clustering, and 0 for the scenario where the clustering provides no information about the true labels.

```python
def plot_confusion_matrix(y_test, y_pred, y_test_label=None, y_pred_label=None):
    """ plot confusion matrix for the predicted labels """

    df = pd.DataFrame({"predicted_label": y_pred, "true_label": y_test})
    ct = pd.crosstab(df['predicted_label'], df['true_label']).T

    kwsHeat = {
        'annot': True,
        'cbar': False,
        'square': True,
        'xticklabels': True,
        'yticklabels': True,
        'linewidths': .1,
        'fmt': 'd',
        'cmap': 'YlGnBu',
    }

    ax = sns.heatmap(ct, **kwsHeat)
    ax.tick_params(axis='both', which='both', labelsize=8)
    ax.set_xlabel('Predicted Label' if y_pred_label is None else y_pred_label, fontweight='bold')
    ax.set_ylabel('True Label' if y_test_label is None else y_test_label, fontweight='bold')
    ax.set_title('Confusion Matrix', fontweight='bold')
```

```python
# get HC cluster index label for each ECG signal
hc_clusters = fcluster(hc_linkage_matrix, hc_optimal_n_of_clusters, criterion='maxclust')
hc_clusters_str = [str(cluster_id) for cluster_id in hc_clusters]
```

```python
# Plot confusion matrices
plt.figure(figsize=(10, 7))
plot_confusion_matrix(y_test=true_labels, y_pred=hc_clusters_str, y_test_label='ECG True Label', y_pred_label='Predicted HC Cluster Index')
```

```python
# Adjusted Rand Index (ARI)
ari = adjusted_rand_score(true_labels, hc_clusters_str)
print("Adjusted Rand Index (ARI):", ari)

# Normalized Mutual Information (NMI)
nmi = normalized_mutual_info_score(true_labels, hc_clusters_str)
print("Normalized Mutual Information (NMI):", nmi)
```
