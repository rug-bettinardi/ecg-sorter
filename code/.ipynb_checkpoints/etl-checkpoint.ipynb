# Create Meta

### Meta helper functions:

```python
def aggregate_diagnostic(y_dic):
    """
    Aggregate diagnostic classes based on keys from a dictionary that exist in a global DataFrame.

    This function iterates over keys in the input dictionary and aggregates diagnostic classes
    from a predefined DataFrame `agg_df` where keys match indices. The function returns a list
    of unique diagnostic classes.

    Args:
        y_dic (dict): A dictionary where keys represent some identifiers which are used to look up
                      diagnostic classes in a global DataFrame `agg_df`.

    Returns:
        list: A list of unique diagnostic classes corresponding to the keys in the input dictionary
              that are present in `agg_df`.

    """

    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))


def create_multi_label_one_hot_mapping(classes):
    """
    Creates a dictionary mapping class labels to multi-label one-hot vectors.

    Args:
      classes: A list of class labels (strings).

    Returns:
      A dictionary mapping class labels (strings) to one-hot encoded vectors (numpy arrays).
    """

    num_classes = len(classes)
    mapping = {}

    # UNKNOWN class gets all zeros
    mapping['UNKNOWN'] = np.zeros(num_classes, dtype=np.float32)

    # Individual classes get one-hot encoding
    for i, cls in enumerate(classes):
      one_hot = np.zeros(num_classes, dtype=np.float32)
      one_hot[i] = 1.0
      mapping[cls] = one_hot

    # Combinations of classes (use recursion or iteration)
    for i in range(2, len(classes) + 1):
      for combo in itertools.combinations(classes, i):
        combo_name = '_'.join(combo)
        combo_vector = np.sum([mapping[cls] for cls in combo], axis=0)
        mapping[combo_name] = combo_vector

    return mapping



def diagnostic_list_to_string(class_list, mapping):
    """
    Converts a list of diagnostic superclasses to a string based on a mapping dictionary.

    Args:
      class_list: A list of strings representing diagnostic superclasses.
      mapping: A dictionary mapping combinations of superclasses to one-hot encoded vectors.

    Returns:
      A string representing the combination of superclasses based on the mapping dictionary.
    """

    if len(class_list) == 0:
      return "UNKNOWN"

    for superstring in mapping.keys():
      if all([cls in superstring.split('_') for cls in class_list]):
        return superstring


def keep_only_unique_diagnostic_superclasses(df):

  # Apply the function to the diagnostic superclass column
  df['diagnostic_superclass'] = df['diagnostic_superclass'].apply(filter_diagnostic)

  # Filter out rows where diagnostic superclass is None
  df.dropna(subset=['diagnostic_superclass'], inplace=True)

  return df



def filter_diagnostic(diagnostic_str):
  diagnostic_to_keep = ['NORM', 'MI', 'STTC', 'CD', 'HYP']

  if diagnostic_str in diagnostic_to_keep:
    return diagnostic_str

  else:
    try:
        # Evaluate the string to a list
        diagnostic_list = ast.literal_eval(diagnostic_str)
        # Check if it's a single-element list and if the element is in the valid list
        if len(diagnostic_list) == 1 and diagnostic_list[0] in diagnostic_to_keep:
            return diagnostic_list[0]
        else:
            return None
    except:
        return None
```

### Load raw csv and create complete Meta

```python
# load meta as dataframe
meta = pd.read_csv('data/ptbxl/raw/ptbxl_database.csv', index_col='ecg_id')
meta.scp_codes = meta.scp_codes.apply(lambda x: ast.literal_eval(x))

# Load scp_statements.csv for diagnostic aggregation
agg_df = pd.read_csv('data/ptbxl/raw/scp_statements.csv', index_col=0)
agg_df = agg_df[agg_df.diagnostic == 1]

# Apply diagnostic superclass
meta['diagnostic_superclass'] = meta.scp_codes.apply(aggregate_diagnostic)

# save as .csv
meta.to_csv('data/ptbxl/ptbxl_meta_complete.csv')
```

### Create reduced, balanced Meta

```python
def reduce_normals(df, label_column='diagnostic_label', norm_label='NORM', max_normals=4000):

    # Separate 'NORM' and non-'NORM' rows
    norm_rows = df[df[label_column] == norm_label]
    other_rows = df[df[label_column] != norm_label]

    # Randomly sample the desired number of 'NORM' rows
    sampled_norm_rows = norm_rows.sample(n=max_normals, random_state=42)  # Setting random_state for reproducibility

    # Combine the sampled 'NORM' rows with the other rows
    reduced_df = pd.concat([sampled_norm_rows, other_rows])

    return reduced_df
```

```python
# create redux version (keep only records with 1 diagnostic superclass)
meta['num_diagnostic_superclasses'] = meta['diagnostic_superclass'].apply(lambda x: len(x))
meta_redux = meta.query("num_diagnostic_superclasses == 1")
meta_redux['diagnostic_label'] = meta_redux['diagnostic_superclass'].apply(lambda x: x[0])
meta_redux_balanced = reduce_normals(meta_redux, max_normals=4000)
meta_redux_balanced.to_csv('data/ptbxl/ptbxl_meta_redux_balanced.csv')
```

# Create TFRecord Dataset

Please note that creating from scratch a TFRecord Dataset takes time. Creating a dataset from the full database (~21k+) took 2h30.

Define TFRecord helper functions

```python
def create_tfrecord_example(signal, label):
  """
  Create a TFRecord example from the signal data and label.

  Args:
    signal (np.ndarray): The signal data array (float64).
    label (str): The label corresponding to the signal data (string).

  Returns:
    tf.train.Example: The TFRecord example.
  """

  # prepare signal
  signal = signal.astype(np.float32)
  signal = signal.flatten()

  # prepare label
  label = label.encode('utf-8')

  feature = {
      'signal': tf.train.Feature(float_list=tf.train.FloatList(value=signal)),
      'label': tf.train.Feature(bytes_list=tf.train.BytesList(value=[label])),
  }

  example = tf.train.Example(features=tf.train.Features(feature=feature))

  return example



def parse_signal_and_label(example):
  """
  Parse a TFRecord example into signal and label tensors.

  Args:
      proto: A serialized tf.train.Example.

  Returns:
      signal: A tf.Tensor representing the signal data (float32).
      label: A tf.Tensor representing the label (string).
  """

  feature_description = {
      'signal': tf.io.VarLenFeature(tf.float32),
      'label': tf.io.FixedLenFeature([], dtype=tf.string),
  }

  parsed_features = tf.io.parse_single_example(example, feature_description)

  # Decode and reshape the signal and label appropriately.
  signal = tf.sparse.to_dense(parsed_features['signal'])
  signal = tf.reshape(signal, shape=[1000, 12])

  label = parsed_features['label']

  return signal, label



def parse_signal_and_signal(example):
  """
  Parse a TFRecord example into signal and signal tensors, needed as input
  for autoencoder

  Args:
      proto: A serialized tf.train.Example.

  Returns:
      signal: A tf.Tensor representing the signal data (float32).
      signal: A tf.Tensor representing the signal data (float32).
  """

  feature_description = {
      'signal': tf.io.VarLenFeature(tf.float32),
  }

  parsed_features = tf.io.parse_single_example(example, feature_description)

  # Decode and reshape the signal and label appropriately.
  signal = tf.sparse.to_dense(parsed_features['signal'])
  signal = tf.reshape(signal, shape=[1000, 12])

  return signal, signal



def dask_load_raw_data_to_tfrecords(df, path, output_dir, num_shards, num_files_to_load=None):
  """
  Load raw signal data using dask parallelization, optionally selecting only the first N files,
  and save the data to multiple TFRecords files.

  Args:
      df (DataFrame): A pandas DataFrame containing the file names under columns
                      'filename_lr' for low resolution and 'filename_hr' for high resolution.
      path (str): The file path prefix required to access the data files.
      output_dir (str): The directory where the TFRecords files will be saved.
      num_shards (int): The number of TFRecords shards to create.
      num_files_to_load (int, optional): The number of files to select and load. If not specified, all files are loaded.

  Returns:
      None
  """

  file_list = df['filename_lr'].tolist()

  # Ensure the output directory exists
  if not os.path.exists(output_dir):
      os.makedirs(output_dir)

  # Extract labels from the DataFrame
  labels = df['diagnostic_label'].values

  # If num_files_to_load is specified, select only the first num_files_to_load files
  if num_files_to_load is not None:
      file_list = file_list[:num_files_to_load]
      labels = labels[:num_files_to_load]

  @delayed
  def load_file(file):
      # Load the signal data from the file
      signal, meta = wfdb.rdsamp(os.path.join(path, file))
      return signal

  # Create delayed tasks for loading data from the selected files
  delayed_results = [load_file(f) for f in file_list]

  # Load data from the selected files with progress bar
  with tqdm(total=len(file_list), desc="Loading data files") as pbar:
      results = compute(*delayed_results, scheduler='threads')
      for _ in results:
          pbar.update(1)

  # Convert the list of results to a numpy array
  signals = np.array(results)

  # Split data and labels into shards
  shard_size = len(signals) // num_shards
  for shard_id in range(num_shards):
      start_idx = shard_id * shard_size
      end_idx = (shard_id + 1) * shard_size if shard_id != num_shards - 1 else len(signals)
      shard_signals = signals[start_idx:end_idx]
      shard_labels = labels[start_idx:end_idx]
      shard_filename = os.path.join(output_dir, f"data_{shard_id:04d}.tfrec")

      # Write data to TFRecords shard
      with tf.io.TFRecordWriter(shard_filename) as writer:
          for i in range(len(shard_signals)):
              example = create_tfrecord_example(shard_signals[i], shard_labels[i])
              writer.write(example.SerializeToString())

  print(f"FINISHED: TFRecords shards created at: {output_dir}")
```

Load and store signal data and labels in `TFRecord` format:

```python
dask_load_raw_data_to_tfrecords(
    df=meta_redux_balanced,
    path='data/ptbxl/raw',
    output_dir='data/ptbxl/tfrecords_redux_balanced',
    num_shards=10,
    num_files_to_load=None
    )
```
